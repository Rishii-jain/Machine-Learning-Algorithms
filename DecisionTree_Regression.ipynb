```python
import pandas as pd
from pandas_datareader import data
import numpy as np
from  sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df= pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
model = DecisionTreeRegressor(criterion = 'friedman_mse', max_depth = 5)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

HyperParameter Tuning

```python
param_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'max_depth': [None, 1, 2, 3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [0.25, 0.5,1]
}
```

```python
reg = GridSearchCV(DecisionTreeRegressor(), param_grid)
```

```python
reg.fit(x_train, y_train)
```

```python
r2_score(y_test, reg.predict(x_test))
```

```python
for importance, name in zip(reg.best_estimator_.feature_importances_, iris.feature_names):
    print(name, importance)
```
